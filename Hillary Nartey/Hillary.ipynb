```python
from urllib.request import Request,urlopen
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
url = 'https://www.jumia.com.gh/laptops/'
url
```

```python
request = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
request
```

```python
webpage = urlopen(request).read()
```

```python
formatted_webpage = soup(webpage)
formatted_webpage
```

```python
#get element by class
results = formatted_webpage.select(".info .name")
results
```

```python
formatted_webpage = soup(webpage, "html.parser")
```

```python
products = [pro.get_text() for pro in formatted_webpage .select(".info .name")]
```

```python
price = [cost.get_text() for cost in formatted_webpage .select(".info .prc")]
```

```python
df = pd.DataFrame({
    "name": products,
    "price": price,})
df
```

```python
df.to_csv('hillary_results.csv')
```
