```python
import pandas as pd
from selenium import webdriver
```

```python
url = 'https://www.jumia.com.gh/laptops/'
path="C:/webdriver/chromedriver.exe"
```

```python
driver = webdriver.Chrome(path)
```

```python
driver.get(url)
```

```python
names=[name.text for name in driver.find_elements_by_xpath("//h3")]
```

```python
names
```

```python
prices=[price.text for price in driver.find_elements_by_xpath('//div[@class="info"]//child::div[@class="prc"]')]
```

```python
prices
```

```python
links = [link.get_attribute("href") for link in driver.find_elements_by_xpath("//article[contains(@class,'prd _fb')]//child::a") ]
links
```

```python
df= pd.DataFrame({"Name":names,"Price":prices, "Links":links})
```

```python
df
```

```python
df.to_csv("../csv/laptop_prices_links.csv")
```
