```python
from urllib.request import Request,urlopen
# import request and urlopen libraries
```

```python
from bs4 import BeautifulSoup as bs
# import Beautiful Soup library
```

```python
url = 'https://www.jumia.com.gh/laptops/'
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
# send request to url
```

```python
webpage = urlopen(req).read()
webpage
# open url response
```

```python
page_soup = bs(webpage, "html.parser")
```

```python
products = [pro.get_text() for pro in page_soup.select(".info .name")]
# get all product names of the page using the div class= info and h3 class = name
```

```python
price = [cost.get_text() for cost in page_soup.select(".info .prc")]
#  get all prices of products of the page using the div class= info and div class = prc
```

```python
import pandas as pd
# import pandas library
laptops = pd.DataFrame({
    "Laptop name": products,
    "Price": price,
    
})
laptops

# create dataframe with the columns laptop name and price
```

```python
laptops.to_csv("Jumia Laptops Assigment From Godwin Cleland.csv")
```
