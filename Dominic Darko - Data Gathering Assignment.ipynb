```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests as rs
import re
```

## DMC-Natura-XL Specs Scraping

```python
url = 'https://www.yarnplaza.com/product/4216/dmc-natura-xl.html'

page = rs.get(url)

page.content
```

```python
soup = bs(page.content, 'html.parser')
soup
```

```python
tag = soup.find('div', class_ = re.compile(r'innerspecsholder'))
tag
```

```python
tag_2 = tag.find('table')
tag_2
```

### DMC-Natura-XL Final Specs

```python
item_specs = tag_2.find_all('td')
item_specs = [specs.text for specs in item_specs[6:10]]
item_specs
```

## DMC-Natura-XL Price Scraping

```python
tag2 = soup.find('div', class_ = re.compile(r'productlist-bottomfx'))
tag2
```

```python
price_tag = soup.find('div', class_ = 'buy-price')
price_tag
```

### DMC-Natura-XL Final Price

```python
xl_price_tag = soup.find('span', class_ = 'product-price-amount')
buy_price = [xl.text for xl in xl_price_tag]
print(f'Natura XL Price = {buy_price}')
```
